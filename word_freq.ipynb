```python
import csv
import sys
import collections as clt
import re
from itertools import *
train_q =csv.reader(open("C:\Users\leonidf\Documents\Kaggle 2\\training_set.tsv", 'r'), delimiter = '\t')
#train_q =csv.reader(open('training_set.tsv', 'r'))
```

```python
q_all = []
for i in train_q :
    q_all.append(i[1].split())
    q_all.append(i[3].split())
    q_all.append(i[4].split())
    q_all.append(i[5].split())
    q_all.append(i[6].split())
    #
```

```python
q_words = list(chain(*q_all))
```
